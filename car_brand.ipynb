```python
# Gerekli kütüphaneleri yükle
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Eğitilen modeli yükle (yolu senin klasörüne göre ayarlı)
model = YOLO("car_brand_detect/yolov8n_v14/weights/best.pt")

# Görselin tam yolunu doğru ver (örnek: /content/resim.jpg)
image_path = "/content/resim.jpg"

# Tahmin yap (show=False çünkü Colab GUI desteklemez)
results = model.predict(image_path, show=False)

# Görseli RGB formatında oku
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tahmin sonuçlarını çiz
for result in results:
    plotted_img = result.plot()  # prediction box'lu görseli al
    plt.figure(figsize=(10, 8))
    plt.imshow(plotted_img)
    plt.axis("off")
    plt.title("Marka Tespiti Tahmini")
    plt.show()
```
